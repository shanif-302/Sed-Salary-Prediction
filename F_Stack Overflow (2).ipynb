```python
def digest_categories(categories,cutoff):
    categoricalMap = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff : 
            categoricalMap[categories.index[i]] = categories.index[i]
        else:
            categoricalMap[categories.index[i]] = 'Other'

    return categoricalMap
```

```python
countryMap = digest_categories(df.Country.value_counts(),400)
print(countryMap)
```

```python
df.Country = df['Country'].map(countryMap)
```

```python
def setExperience(i):
    
   if i == 'Less than 1 year':
       return 0.5
   if i == 'More than 50 years':
       return 50
   return float (i)
```

```python
df['YearsCode'] = df['YearsCode'].apply(setExperience)
```

```python
df.YearsCode.unique()
```

```python
df.EdLevel.unique()
```

```python
def setEducation(i):
    if 'Master’s degree' in i:
        return 'Master’s degree'
    if 'Bachelor’s degree' in i:
        return 'Bachelor’s degree' 
    if 'Professional degree' in i or 'Other doctoral degree'in i:
        return 'post'
    return 'less  than a Bachelors'
```

```python
df['EdLevel'] = df['EdLevel'].apply(setExperience)
```

```python
df['EdLevel'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LeEducation = LabelEncoder()
df['EdLevel'] = LeEducation.fit_transform(df['EdLevel'])
```

```python
df['Country'].unique()
```

```python
LeCountry = LabelEncoder()
df['Country'] = LeCountry.fit_transform(df['Country'])
```

### Preparation of data to feed the model

```python
X = df.drop('Salary',axis = 1)
y = df['Salary']
```

### Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
model_lr.fit(X,y)
```

```python
y_pred = model_lr.predict(X)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
```

```python
errorright = np.sqrt(mean_squared_error(y,y_pred))
```

```python
print(errorright)
```
